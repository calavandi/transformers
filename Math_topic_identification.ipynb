```python
dataset = load_dataset(
   'math_dataset', 'algebra__polynomial_roots')
# Define the training dataset
train_dataset = dataset['train']

cleaned_dataset = []

for question in train_dataset['question'][:5]:
    cleaned_example = {
         'question': str(question),
         'category_label': "algebra__polynomial_roots"
     }
    cleaned_dataset.append(cleaned_example)


```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from transformers import AdamW, get_linear_schedule_with_warmup
from sentence_transformers import SentenceTransformer, InputExample

model = SentenceTransformer('all-mpnet-base-v2')


batch_size = 16
train_data_loader = DataLoader(cleaned_dataset, batch_size=batch_size, shuffle=True)


criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=2e-5)

num_epochs = 100
output_path = 'trained_model'

for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0

    for batch in train_data_loader:
        questions = batch['question']
        category_labels = batch['category_label']

        optimizer.zero_grad()

        embeddings = model.encode(questions, convert_to_tensor=True)


        category_numerical_label = 0


        labels = torch.tensor([category_numerical_label] * len(questions), dtype=torch.long)


        embeddings.requires_grad = True

        loss = criterion(embeddings, labels)

        loss.backward()
        optimizer.step()


        total_loss += loss.item()

    avg_loss = total_loss / len(train_data_loader)
    print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {avg_loss:.4f}")


model.save(output_path)
```
